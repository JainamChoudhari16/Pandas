## ***Data Cleaning***

```python
#creating a simple day to day temperature table for data cleaning
data = {
    "Day": ["2021-01-01", "2021-01-05", "2021-01-06", "2021-01-07", "2021-01-08", "2021-01-09", "2021-01-10", "2021-01-11", "2021-01-12", "2021-01-13"],
    "Temperature": [34, 33333, 28, 32, 33333, 26, 29, 31, 33333, 35],
    "Wind Speed": [6, 9, 33333, 5, 8, 11111, 33333, 33333, 9, 6],
    "Event": ["Sunny", "Rain", float("nan"), "Rain", float("nan"), "Snow", float("nan"), "Sunny", "Rain", "Rain"]
}
df = pd.DataFrame(data)
df
```

```python
#in the above table there are some inappropriate values and missing data i.e. Nan i.e. null.
#so firstly we will replace inappropriate values with null value i.e. NaN
df1 = df.replace([33333,11111],np.NaN)
df1
```

```python
#checking total number of null values in each column
df1.isnull().sum()
```

```python
#Now to solve the problem of missing(null) values data we have 2 methods 1st is to fill the null value data and 2nd is to remove the null value data(row/column).
```

```python
#i.e. removing rows and columns
#here we use dropna() fuction: this function by default removes all rows where there is atleast 1 null value
df1.dropna()
```

```python
#To remove columns with null value in dropna() function we assign axis value 1
df1.dropna(axis=1)
```

```python
df1.dropna(thresh=3) #this will not remove that rows which have atleast 3 values
```

```python
#Handling the missing data by filling that data
df1.fillna("Hii") #This function will fill Hii in place of all null values
```

```python
df1.fillna(method='ffill') #this will fill previous values in next null values
```

```python
#Merging two or more dataset
df2 = pd.DataFrame({
    "Name":['a','b','c','d','e'],
    "age":[11,12,13,14,15],
    "sub":['s1','s2','s3','s4','s5']
})
df3 = pd.DataFrame({
    "Name":['a','bb','c','d','e'],
    "age":[11,12,13,44,15],
    "sub":['s1','s2','s33','s4','s5']
})

#There are several method of merging the dataframes:
```

```python
#Method 1: Concat method
merged_df = pd.concat([df2,df3],ignore_index=True)
merged_df
```

```python
#Method 2: Merge method which works as join from sql.
merg_df=pd.merge(df2,df3) #by default merge method prints the row that is common in both dataframes i.e. Inner join of sql
merg_df
```

```python
merg_df=pd.merge(left=df2,right=df3,how="right")
merg_df #this merge method prints the row that is in right dataframe i.e. Right join of sql
```

```python
merg_df=pd.merge(left=df2,right=df3,how="left")
merg_df #this merge method prints the row that is in left dataframe i.e. Left join of sql
```

```python
merg_df=pd.merge(left=df2,right=df3,how="outer")
merg_df #this merge method prints all the different rows in both dataframes i.e. Fully Outter join of sql
```

```python
merg_df = pd.merge(df2,df3,on='Name')
merg_df #prints table with rows that have same Name's value in both the table
#In output age_x is age of left table age_y is for right side table
```
